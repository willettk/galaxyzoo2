```python
n = 3
f = symbols('f:%i'%(n+1), real=True, positive=True)
p = symbols('p:%i'%(n+1), real=True, positive=True)
r = [[symbols('r%i%i'%(i, j), real=True, positive=True) for j in range(n+1)] for i in range(n+1)]
ii, jj = symbols('i,j', integer=True, positive=True)
```

```python
def pexpr(i):
    return f[i] - sum(p[jj]*r[jj][i], (jj, 1, i-1)) - sum(p[jj]*r[jj][i], (jj, i+1, n)) - p[i]

pexpr(3)
```
